```python
def factorial(n):
    """Returns n! for n >= 0."""
    result = 1
    for i in range(2, n+1):
        result *= i
    return result

print (factorial("1"))
```

```python
def factorial(n):
    result = 1
    i = 1

    while i <= n:
        result = result * i
        i = i + 1

    return result


print(factorial(0))
```

```python
class Factorial:
    def __init__(self, n):
        self.n = n

    def compute(self):
        result = 1
        i = 1

        while i <= self.n:
            result = result * i
            i = i + 1

        return result


# Example
f = Factorial(3)
print(f.compute())
```

```python
def fibonacci(n):
    fib = []
    a, b = 0, 1

    for i in range(n):
        fib.append(a)
        old_a = a
        a = b 
        b= old_a + b

    return fib


print(fibonacci(7))
```

```python
def zigzag_array(rows, cols):
    num = 1
    #matrix = [[0]*cols for _ in range(rows)]
    matrix = [[0 for col in range(cols)] for row in range(rows)]
    for row in matrix:
        print(row)

    for r in range(rows):
        if r % 2 == 0:  # even row → left to right
            for c in range(cols):
                matrix[r][c] = num
                num += 1
        else:           # odd row → right to left
            for c in range(cols-1, -1, -1):
                matrix[r][c] = num
                num += 1
    return matrix

# Example
res = zigzag_array(3,3)
print ("************* zigzag array****************")
for row in res:
    print(row)
```

```python
def diagonal_zigzag(n):
    #matrix = [[0]*n for _ in range(n)]
    matrix = [[0 for col in range(n)] for row in range(n)]
    num = 1
    
    for s in range(2*n - 1):
        if s % 2 == 0:  # even diagonal → go up
            r = min(s, n-1)
            c = s - r
            while r >= 0 and c < n:
                matrix[r][c] = num
                num += 1
                r -= 1
                c += 1
        else:           # odd diagonal → go down
            c = min(s, n-1)
            r = s - c
            while c >= 0 and r < n:
                matrix[r][c] = num
                num += 1
                r += 1
                c -= 1
    return matrix

# Example
res = diagonal_zigzag(4)
for row in res:
    print(row)
```

```python
def find_max(numbers):
    """Returns the maximum in a list, or None if empty."""
    return max(numbers) if numbers else None

print(find_max([1,2,3]))
```

```python
#************************************
#Find Second Largest Element
#************************************

def find_second_largest(numbers= None):
    """Find second largest element in list."""
    # Handle edge cases
    if numbers is None or len(numbers) < 2:
        return None
    
    # Remove duplicates and sort
    unique_nums = list(set(numbers))
    unique_nums.sort(reverse=True)
    
    if len(unique_nums) < 2:
        return None
    
    return unique_nums[1]

# Test cases
print(find_second_largest([1, 2, 3, 4, 5]))  # 4
```

```python
def is_palindrome(s):
    """Checks if s is a palindrome."""
    s = s.lower()
    return s == s[::-1]
print (is_palindrome("abcba"))
```

```python
def is_palindrome(s):
    s = s.lower()
    rev = ""

    # make reverse string
    for ch in s:
        rev = ch + rev

    return s == rev

print(is_palindrome("abcba"))
```

```python
class PalindromeChecker:
    def __init__(self, text):
        self.text = text.lower()

    def is_palindrome(self):
        return self.text == self.text[::-1]


# Example
word = PalindromeChecker("abcba")
print(word.is_palindrome())
```

```python
def reverse_only_letters(s):
    letters = [ch for ch in s if ch.isalpha()]   # collect only letters
    letters.reverse()                            # reverse the letters

    result = []
    idx = 0

    for ch in s:
        if ch.isalpha():
            result.append(letters[idx])          # place reversed letter
            idx += 1
        else:
            result.append(ch)                    # keep numbers same

    return "".join(result)

# Example
print(reverse_only_letters("a1b2c3d")) 
```

```python
def reverse_words(sentence):
    """Reverses the words in a space-separated string."""
    return ' '.join(sentence.split()[::-1])

print (reverse_words("Sujit is a Boy"))
```

```python
def reverse_words(sentence):
    words = sentence.split()     # split into list of words
    reversed_words = []
    print("word list from the sentaence is as :")
    print(words)
    # manually reverse the list
    for w in words:
        reversed_words.insert(0, w)

    # join back into a string
    return " ".join(reversed_words)


print(reverse_words("Sujit is a Boy"))
```
